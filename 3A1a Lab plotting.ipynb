```python
#Importing necessary modules
import numpy as np
import matplotlib.pyplot as plt

#Calculated variables in order to find Cd and Re
rho_air_plate = 1.161
rho_air_sphere = 1.155
rho_air_tripped_sphere = 1.152
rho_air_streamlined = 1.151
nu_plate = 1.5657e-5
nu_sphere = 1.5784e-5
nu_tripped_sphere = 1.5956e-5
nu_streamlined = 1.6e-5
cross_sectional_area_all = 0.01831
d_all = 0.1527

#Raw data from plate experiment
pressure_difference_plate = np.array([1.01, 2.06, 3.05, 4, 4.98, 5.98, 6.98, 7.76, 8.98, 10.01, 10.98, 11.98]) * 100
counter_weight_plate = np.array([22.1,  45.8, 68.8, 88.9, 111.2, 134.4, 156.2, 173.3, 199.5, 224.1, 243.6, 265.1])
drag_plate = counter_weight_plate * 0.1052
velocity_plate = np.sqrt(2 / rho_air_plate * (pressure_difference_plate))
print(velocity_plate)

#Raw data from sphere experiment
pressure_difference_sphere = np.array([1.02, 2.02, 3.04, 3.98, 4.99,  6.02, 6.98, 7.49, 7.86, 8.67, 9.05, 9.49, 10, 10.94, 12]) * 100
counter_weight_sphere = np.array([11.3, 24.7, 34.6, 44.1, 54.3, 66.1, 74.6, 67.4, 68.7, 67.1, 66, 68.4, 70.4, 73.5, 75.3])
drag_sphere = counter_weight_sphere * 0.1052
velocity_sphere = np.sqrt(2 / rho_air_sphere * (pressure_difference_sphere))

#Raw data from tripped sphere experiment
pressure_difference_tripped_sphere = np.array([1.02, 1.98, 2.98, 4.03, 4.95, 5.47, 6.02, 6.51, 6.98, 7.51, 7.9, 8.68, 9.05, 10.06, 10.98, 12.03]) * 100
counter_weight_tripped_sphere = np.array([8.9, 17.7, 26.4, 32, 36.5, 39.8, 43.5, 46.1, 49.2, 53.4, 55.2, 58.6, 62, 67.6, 73.7, 75.6])
drag_tripped_sphere = counter_weight_tripped_sphere * 0.1052
velocity_tripped_sphere = np.sqrt(2 / rho_air_tripped_sphere * (pressure_difference_tripped_sphere))

#Raw data from streamlined experiment
pressure_difference_streamlined = np.array([0.98, 2, 3, 3.95, 4.99, 6.01, 7.02, 7.92, 9.03, 9.96, 11, 11.98]) * 100
counter_weight_streamlined = np.array([5.1, 9.5, 13.7, 16.7, 20.9, 25.1, 28.8, 33.9, 36.4, 37.1, 41.6, 44.2])
drag_streamlined = counter_weight_streamlined * 0.1052
velocity_streamlined = np.sqrt(2 / rho_air_streamlined * (pressure_difference_streamlined))

#Cd and Re arrays
cd_plate = drag_plate / (0.5 * rho_air_plate * velocity_plate * velocity_plate * cross_sectional_area_all) - 0.1
re_plate = velocity_plate * d_all / nu_plate

cd_sphere = drag_sphere / (0.5 * rho_air_sphere * velocity_sphere ** 2 * cross_sectional_area_all) - 0.1
re_sphere = velocity_sphere * d_all / nu_sphere

cd_tripped_sphere = drag_tripped_sphere / (0.5 * rho_air_tripped_sphere * velocity_tripped_sphere ** 2 * cross_sectional_area_all) - 0.1
re_tripped_sphere = velocity_tripped_sphere * d_all / nu_tripped_sphere

cd_streamlined = drag_streamlined / (0.5 * rho_air_streamlined * velocity_streamlined ** 2 * cross_sectional_area_all) - 0.1
re_streamlined = velocity_streamlined * d_all / nu_streamlined

#Plotting Cd against Re for the plate
plt.plot(re_plate, cd_plate, color='#0F4722')
plt.annotate("Flat Plate", (445000, 1.15), color='#0F4722')
plt.plot(re_sphere, cd_sphere, color='#157C39')
plt.annotate("Sphere", (443000, 0.2580), color='#0F4722')
plt.plot(re_tripped_sphere, cd_tripped_sphere, color='#18BB51')
plt.annotate("Tripped Sphere", (443000, 0.21), color ='#18BB51')
plt.plot(re_streamlined, cd_streamlined, color='#6CF59C')
plt.annotate("Streamlined Body", (440000, 0.1), color = '#6CF59C')
#plt.axvline(x = 348000, color = 'black', label = 'Critical Re number', alpha = 0.25)
plt.xlabel('Reynolds Number')
plt.ylabel('Drag Coefficient')
#plt.plot(348000, 0.4180, marker = 'X')
#plt.text(344000, 0.4100, "Critical Reynolds Number", size = 8, horizontalalignment='right', alpha = 0.8)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(linestyle = '--', linewidth = 0.1)
plt.show()

plt.plot(re_plate, cd_plate, color='#0F4722')
plt.annotate("Flat Plate", (445000, 1.15), color='#0F4722')
plt.plot(re_sphere, cd_sphere, color='#157C39')
plt.annotate("Sphere", (443000, 0.2580), color='#0F4722')
plt.plot(re_tripped_sphere, cd_tripped_sphere, color='#18BB51')
plt.annotate("Tripped Sphere", (443000, 0.21), color ='#18BB51')
plt.plot(re_streamlined, cd_streamlined, color='#6CF59C')
plt.annotate("Streamlined Body", (440000, 0.1), color = '#6CF59C')
plt.axvline(x = 348000, color = 'black', label = 'Critical Re number', alpha = 0.25)
plt.xlabel('Reynolds Number')
plt.ylabel('Drag Coefficient')
plt.plot(348000, 0.4180, marker = 'X')
plt.text(344000, 0.4100, "Critical Reynolds Number", size = 8, horizontalalignment='right', alpha = 0.8)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(linestyle = '--', linewidth = 0.1)
plt.show()

plt.plot(re_plate, cd_plate, color='#0F4722')
plt.annotate("Flat Plate", (445000, 1.15), color='#0F4722')
plt.errorbar(re_plate[0], cd_plate[0], yerr = 0.00787 * cd_plate[0])
plt.plot(re_sphere, cd_sphere, color='#157C39')
plt.annotate("Sphere", (443000, 0.2580), color='#0F4722')
plt.errorbar(re_sphere[0], cd_sphere[0], yerr = 0.00998 * cd_sphere[0])
plt.plot(re_tripped_sphere, cd_tripped_sphere, color='#18BB51')
plt.annotate("Tripped Sphere", (443000, 0.21), color ='#18BB51')
plt.errorbar(re_tripped_sphere[0], cd_tripped_sphere[0], yerr = 0.01118 * cd_tripped_sphere[0])
plt.plot(re_streamlined, cd_streamlined, color='#6CF59C')
plt.annotate("Streamlined Body", (440000, 0.1), color = '#6CF59C')
plt.errorbar(re_streamlined[0], cd_streamlined[0], yerr = 0.01556 * cd_streamlined[0])
#plt.axvline(x = 348000, color = 'black', label = 'Critical Re number', alpha = 0.25)
plt.xlabel('Reynolds Number')
plt.ylabel('Drag Coefficient')
#plt.plot(348000, 0.4180, marker = 'X')
#plt.text(344000, 0.4100, "Critical Reynolds Number", size = 8, horizontalalignment='right', alpha = 0.8)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(linestyle = '--', linewidth = 0.1)
plt.show()

plt.loglog(re_plate, cd_plate, color='#0F4722')
plt.loglog(re_sphere, cd_sphere, color='#0F4722')
plt.loglog(re_tripped_sphere, cd_tripped_sphere, color ='#18BB51')
plt.loglog(re_streamlined, cd_streamlined, color = '#6CF59C')
plt.xlabel('Reynolds Number')
plt.ylabel('Drag Coefficient')
plt.grid(linestyle = '--', linewidth = 0.1)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend(["Flat Plate", "Sphere", "Tripped sphere", "Streamlined"], bbox_to_anchor=(1.02, 0.1), loc='upper left')
plt.show()

```
