```python
import matplotlib.pyplot as plt
import numpy as np

def l(x):
  return 25/41 * x + 10/41 * x**2 + 6/41 * x**3

def rho(x):
  return x**5


e = 0.325
p_0 = e
t = np.linspace(0, 999, 1000)
p = np.linspace(p_0, p_0, 1000)
for i in range(1, len(p)):
  p[i] = e * l(1 - rho(1 - p[i-1]))

plt.plot(t, p, label='epsilon = 0.325')


e = 0.3275
p_0 = e
t = np.linspace(0, 999, 1000)
p = np.linspace(p_0, p_0, 1000)
for i in range(1, len(p)):
  p[i] = e * l(1 - rho(1 - p[i-1]))

plt.plot(t, p, label='epsilon = 0.33')

e = 0.328
p_0 = e
t = np.linspace(0, 999, 1000)
p = np.linspace(p_0, p_0, 1000)
for i in range(1, len(p)):
  p[i] = e * l(1 - rho(1 - p[i-1]))

plt.plot(t, p, label='epsilon = 0.33')

e = 0.33
p_0 = e
t = np.linspace(0, 999, 1000)
p = np.linspace(p_0, p_0, 1000)
for i in range(1, len(p)):
  p[i] = e * l(1 - rho(1 - p[i-1]))

plt.plot(t, p, label='epsilon = 0.33')


plt.xlim(0, 500)
plt.grid()
plt.legend(loc='lower right')
plt.show()

```
