```python
#Question 1
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

w = np.linspace(-10000, 10000, 100000)
x = 1.18 * np.cos(0.38*w)
y = -1.18 * np.sin(0.38*w)

plt.plot(x, y)
plt.axis("equal")
plt.grid()
plt.show()
```

```python
#a Nyquist diagram is just an argand diagram.
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

w = np.linspace(-10000, 10000, 100000)
T = 0.01
x = -2 / (1+w**2 * T**2)
y = (-2*w*T)/(1+w**2*T**2)

plt.plot(x, y)
plt.axis("equal")
plt.grid()
plt.show()
```
