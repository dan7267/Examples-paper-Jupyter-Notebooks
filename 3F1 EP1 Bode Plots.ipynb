```python
#Plot 1
from scipy import signal
import matplotlib.pyplot as plt

sys = signal.TransferFunction([1, 2], [1, -1], dt = 0.5)
w, mag, phase = sys.bode()

plt.figure()
plt.semilogx(w, mag)
plt.figure()
plt.semilogx(w, phase)
plt.xlim(0.1, 10)
plt.ylim(-180, 180)
plt.show()
```
