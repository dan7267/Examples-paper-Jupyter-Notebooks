```python
#Question 3(a)
import matplotlib.pyplot as plt
import numpy as np

def psi(r, theta):
  return r**(6/5)*np.sin(6*theta/5)

r = np.linspace(0, 10, 30)
theta = np.linspace(0, 5*np.pi/3, 30)

R, THETA = np.meshgrid(r, theta)
Z = psi(R, THETA)

fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))
cax = ax.contourf(THETA, R, Z, 10)
```

```python
#Question 13(a)
import numpy as np
import matplotlib.pyplot as plt

phi = np.linspace(-3., 24., 51)
psi = np.linspace(-np.pi, np.pi, 21)

F = np.zeros((len(phi), len(psi)), dtype = 'complex')

for i in range(len(phi)):
  for k in range(len(psi)):
    F[i][k] += phi[i]
    F[i][k] += psi[k] * 1j

z = F - np.exp(-F)
plt.plot(np.real(z), np.imag(z))

```

```python
#Question 16 - m=2piuh
import numpy as np
import matplotlib.pyplot as plt

u = 5
h = 1
m = 2 * np.pi * u * h

def psi(x, y):
  return u * y - m / (2 * np.pi) *(np.arctan((y+h)/x) + np.arctan((y-h)/x))

x = np.linspace(-5, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
PSI = psi(X, Y)

levels = np.linspace(-2*np.pi, 4*np.pi, 31)

plt.contour(X, Y, PSI, levels)
plt.show()

```
